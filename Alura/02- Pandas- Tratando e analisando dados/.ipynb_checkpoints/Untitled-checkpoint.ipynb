# Aula 1

## Importando base de dados

```python
import pandas as pd
```

```python
alugueis= pd.read_csv('Dados/aluguel.csv')
alugueis.head()
```

As colunas não foram separadas com a leitura do csv, portanto devemos mudar o separador

```python
alugueis= pd.read_csv('Dados/aluguel.csv', sep=';') # Adicionado parâmetro para mudança do separador padrão ','
alugueis.head()
```

```python
type(alugueis) # Verificação do tipo de dados da minha variável
```

```python
alugueis.info() # Forma mais completa de verificação geral dos dados
```

# Aula 2

## Informações gerais sobre a base de dados

```python
alugueis.dtypes
```

```python
tipos_de_dados= pd.DataFrame(alugueis.dtypes) # Gerando df a partir do .dtypes
tipos_de_dados.columns= ['Tipos de dados'] # Renomeando a coluna de valores
tipos_de_dados.columns.name= 'Variáveis'  # Nomeação da coluna index
tipos_de_dados
```

```python
alugueis.shape # Informa a 'forma' do dataframe LxC (linhas e colunas)
```

```python
print(f'A base de dados apresenta {alugueis.shape[0]} registros de imóveis e {alugueis.shape[1]} variáveis.')
```
